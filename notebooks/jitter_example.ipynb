# Jitter examples

Playground to analyze jitter functions (unrelated to MD Simulations)

```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

import numpy as np
from matplotlib import pyplot as plt
```

# Define jitter function to test

```python
import analyzers.loss as loss_analyzer

fn_jitter = loss_analyzer.get_loss_jitter
```

# Tests

### General test

```python
some_data = np.random.uniform(500, 550, 1000)
x = np.arange(len(some_data))

jitter = fn_jitter(some_data)
print(f"Jitter: {jitter}")
print(f"Mean: {np.mean(some_data)}")

plt.plot(x, some_data)
plt.show()
```

### Test if jitter is independend of function mean value

```python
some_data = np.random.uniform(200, 250, 1000)
x = np.arange(len(some_data))

jitter = fn_jitter(some_data)
print(f"Jitter: {jitter}")
print(f"Mean: {np.mean(some_data)}")

plt.plot(x, some_data)
plt.show()
```

### Test if jitter is proportional to value range

```python
some_data = np.random.uniform(200, 300, 1000)
x = np.arange(len(some_data))

jitter = fn_jitter(some_data)
print(f"Jitter: {jitter}")
print(f"Mean: {np.mean(some_data)}")

plt.plot(x, some_data)
plt.show()
```

### Test if jitter independend of the total number of values

```python
some_data = np.random.uniform(200, 250, 10000)
x = np.arange(len(some_data))

jitter = fn_jitter(some_data)
print(f"Jitter: {jitter}")
print(f"Mean: {np.mean(some_data)}")

plt.plot(x, some_data)
plt.show()
```

### Test if jitter is stable for low number of values

```python
def get_jitter():
    some_data = np.random.uniform(200, 250, 32)
    return fn_jitter(some_data)

jitters = [get_jitter() for _ in range(5000)]
print(f"Mean: {np.mean(jitters)}")
print(f"std: {np.std(jitters)}")
```

### Test if jitter is independend of drift

```python
some_data = np.random.uniform(200, 250, 5000) + np.linspace(0, 1000, 5000)
x = np.arange(len(some_data))

jitter = fn_jitter(some_data)
print(f"Jitter: {jitter}")
print(f"Mean: {np.mean(some_data)}")

plt.plot(x, some_data)
plt.show()
```
