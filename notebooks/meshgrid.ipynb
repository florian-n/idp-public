```python
import numpy as np
import matplotlib.pyplot as plt

from helpers import plotters, grid, converters
```

```python
N_POINTS = 9
N_MOLECULES = N_POINTS ** 3
BOX_SIZE = converters.get_box_length_from_density(N_MOLECULES * 3 , 0.1)

print(f'N_MOLECULES: {N_MOLECULES}')
print(f'BOX_SIZE: {BOX_SIZE}')
```

```python
system = grid.get_points_on_grid(BOX_SIZE, (N_POINTS, N_POINTS, N_POINTS))

print(system.shape)
```

```python
X = system[:, 0]
Y = system[:, 1]
plt.scatter(X, Y)
plt.xticks([0, BOX_SIZE])
plt.yticks([0, BOX_SIZE])
plt.gca().set_aspect('equal')
plt.show()
```

```python
x = [point[0] for point in system]
y = [point[1] for point in system]
z = [point[2] for point in system]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(x, y, z)

ax.set_xlabel(f"{N_POINTS} molecules")
ax.set_ylabel(f"{N_POINTS} molecules")
ax.set_zlabel(f"{N_POINTS} molecules")

plt.savefig("../plots/3d_grid.png")
plt.show()
```
