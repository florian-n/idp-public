```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_loss_jitter(losses: np.array) -> float:
    """
    Calculate the jitter of the losses.
    """
    m, x = np.polyfit(range(len(losses)), losses, 1)
    regression_norm = m * np.arange(len(losses)) + x

    return np.std(losses - regression_norm)
```

```python
N_DATAPOINTS = 32
distribution = np.random.uniform(200, 250, N_DATAPOINTS) + np.linspace(0, 50, N_DATAPOINTS)


m, x = np.polyfit(range(len(distribution)), distribution, 1)
regression_norm = m * np.arange(len(distribution)) + x

plt.plot(distribution, label="Sample distribution")
plt.plot(regression_norm, label="Regression line", c="r")
plt.gcf().set_size_inches(6, 1.6)
plt.grid()
plt.legend(loc="upper left")
plt.savefig("../plots/lossjitter_explaination_1.svg")
plt.show()

new_distribution = distribution - regression_norm
m, x = np.polyfit(range(len(new_distribution)), new_distribution, 1)
regression_norm = m * np.arange(len(new_distribution)) + x

plt.plot(new_distribution, label="Sample distribution (normalized)")
plt.plot(regression_norm, label="Regression line", c="r")
#plt.plot([0, len(distribution)], [0, 0], label="Regression line", c="r")
plt.gcf().set_size_inches(6, 1.6)
plt.legend(loc="upper left")
plt.grid()
plt.yticks(np.linspace(-35, 35, 5))
plt.savefig("../plots/lossjitter_explaination_2.svg")
plt.show()
```
